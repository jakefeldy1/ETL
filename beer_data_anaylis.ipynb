```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import datetime
from datetime import datetime
import numpy 
import os
```

```python
try:
    __thisdir__ = os.path.dirname(os.path.abspath(__file__))
except NameError as error:
    __thisdir__ = "."
    
#create the direcotry to store the data if not existed
data_process = os.path.join(__thisdir__, "less_big.csv")
beer_data_df = pd.read_csv(data_process)
```

```python
beer_data_df.head()
```

```python
#Data clean up 
```

```python
beer_data_cleaned = beer_data_df.drop(columns=['beer_style_id','beer_new_name','brewery_id','beer_beerid'])
```

```python
beer_data_cleaned.head()
```

```python
ber_brewery_rev_df = beer_data_cleaned[["brewery_name","review_overall","beer_name"]]
```

```python
brewries_df = ber_brewery_rev_df.groupby("brewery_name").count()
```

```python
len(brewries_df)
```

```python
brewries_df.sort_values(by=['review_overall'],ascending=False).head(20)
```

```python
beer_df = ber_brewery_rev_df.groupby("beer_name").count()
len(beer_df)
```

```python
top_100_beer_rev = beer_df.sort_values(by=['review_overall'],ascending=False).head(100)
```

```python
beer_rev_df = ber_brewery_rev_df.groupby("beer_name").mean()
```

```python
beer_review_count = (top_100_beer_rev[["brewery_name"]]).rename(columns={'brewery_name': 'review_count'})
```

```python
beer_review_count.head()
```

```python
top_100_beers = pd.merge(beer_review_count,beer_rev_df,how="inner",left_index=True, right_index=True)
```

```python
top_100_beers.head()
```

```python
beer_brewrey_df= ber_brewery_rev_df[["brewery_name","beer_name"]].drop_duplicates()
beer_brewrey_df.set_index("beer_name", inplace=True)
```

```python
beer_brewrey_df.head()
```

```python
top_100_beers = pd.merge(top_100_beers,beer_brewrey_df,how="inner",left_index=True, right_index=True)
```

```python
top_100_beers.sort_values(by=['review_overall'],ascending=False, inplace=True)
top_100_beers.head()
```

```python
beers_by_brewery = (ber_brewery_rev_df[["brewery_name","beer_name"]].drop_duplicates()).groupby("brewery_name").count()
```

```python
beers_by_brewery.rename(columns={'beer_name': 'beers_released'},inplace=True)
beers_by_brewery.sort_values(by=['beers_released'],ascending=False,inplace=True)
beers_by_brewery.head()
```

```python
test = pd.merge(top_100_beers,beers_by_brewery,on="brewery_name",right_index=True,how="inner")
```
