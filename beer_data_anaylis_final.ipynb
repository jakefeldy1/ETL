```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import datetime
from datetime import datetime
import numpy 
import os
```

```python
try:
    __thisdir__ = os.path.dirname(os.path.abspath(__file__))
except NameError as error:
    __thisdir__ = "."
    
#create the direcotry to store the data if not existed
data_process = os.path.join(__thisdir__, "less_big.csv")
beer_data_df = pd.read_csv(data_process)
```

```python
beer_data_df.head()
```

```python
#Data clean up 
```

```python
beer_data_cleaned = beer_data_df.drop(columns=['beer_style_id','beer_new_name','brewery_id','beer_beerid'])
```

```python
beer_data_cleaned.head()
```

```python
beer_rev_col_list = ["brewery_name",
                     "beer_name",
                     "review_overall",
                     "review_aroma",
                     "review_appearance",
                     "review_palate",
                     "review_taste"]

ber_brewery_rev_df = beer_data_cleaned[beer_rev_col_list]
ber_brewery_rev_df
```

```python
brewries_df = ber_brewery_rev_df.groupby("beer_name").count()
brewries_df.head()
```

```python
len(brewries_df)
```

```python
brewries_df.sort_values(by=['review_overall'],ascending=False).head(20)
```

```python
beer_df = ber_brewery_rev_df.groupby("beer_name").count()
len(beer_df)
```

```python
top_beer_rev = beer_df.sort_values(by=['review_overall'],ascending=False)
top_beer_rev.head()
```

```python
beer_rev_df = ber_brewery_rev_df.groupby("beer_name").mean()
beer_rev_df.head()
```

```python
beer_review_count = (top_beer_rev[["brewery_name"]]).rename(columns={'brewery_name': 'review_count'})
```

```python
beer_review_count.head()
```

```python
top_beer_rev = pd.merge(beer_review_count,beer_rev_df,how="inner",left_index=True, right_index=True)
```

```python
top_beer_rev.head()
```

```python
#Need to update here
list_of_columns = ["brewery_name","beer_name"]

beer_brewrey_df= ber_brewery_rev_df[list_of_columns].drop_duplicates()
beer_brewrey_df.set_index("beer_name", inplace=True)
```

```python
beer_brewrey_df.head()
```

```python
top_beer_rev = pd.merge(top_beer_rev,beer_brewrey_df,how="inner",left_index=True, right_index=True)
```

```python
top_beer_rev.sort_values(by=['review_overall'],ascending=False, inplace=True)
top_beer_rev.head()
```

```python
beers_by_brewery = (ber_brewery_rev_df[["brewery_name","beer_name"]].drop_duplicates()).groupby("brewery_name").count()
beers_by_brewery.head()
```

```python
beers_by_brewery.rename(columns={'beer_name': 'beers_released'},inplace=True)
beers_by_brewery.sort_values(by=['beers_released'],ascending=False,inplace=True)
beers_by_brewery.head()
```

```python
beer_and_brewery_df = pd.merge(top_beer_rev,beers_by_brewery,on="brewery_name",right_index=True,how="inner")
```

```python
beer_and_brewery_df
```

```python
#Need to update here
style_columns = ["beer_name",
                "beer_style",
                "ibu_Min",
                "ibu_Max",
                "abv_Min",
                "abv_Max",
                "srm_Min",
                "srm_Max",
                "og_Min",
                "fg_Min",
                "fg_Max"]

beer_style_df= beer_data_cleaned[style_columns].drop_duplicates()
beer_style_df.set_index("beer_name", inplace=True)
```

```python
beer_style_df.head()
```

```python
ibu_avg = (beer_style_df["ibu_Min"] + beer_style_df["ibu_Max"])/2
srm_avg = round((beer_style_df["srm_Min"] + beer_style_df["srm_Max"])/2)
fg_avg = (beer_style_df["fg_Min"] + beer_style_df["fg_Max"])/2
```

```python
beer_style_df["ibu_avg"] = ibu_avg
beer_style_df["srm_avg"] = srm_avg
beer_style_df["fg_avg"] = fg_avg
beer_style_df.drop(columns=['ibu_Min','srm_Min','ibu_Max','srm_Max',"fg_Min","fg_Max"],inplace=True)
beer_style_df
```

```python
#We need to attach this to the final DF
beer_brewery_style_df = pd.merge(beer_and_brewery_df,beer_style_df,left_index=True, right_index=True,how="inner")
beer_brewery_style_df
```

```python
#Preparing the data to plot beer styles
beer_style_data = beer_brewery_style_df.groupby("beer_style").mean()
```

```python
beer_style_data.drop(columns=["review_count","beers_released"],inplace=True)
beer_style_data.sort_values(by=['review_overall'],ascending=False,inplace=True)
#Push this data for plotting styles reviews
beer_style_data
```

```python
 beer_count_array = (beer_brewery_style_df.groupby("beer_style").count())["review_count"]
```

```python
beer_style_data["review_count"]=beer_count_array
```
